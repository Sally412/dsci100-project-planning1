### 1) Data Description

- **Dataset overview**
  - Players dataset contains 196 observations and 7 variables.  
  - Sessions dataset contains 1535 observations and 5 variables.  
  - Data collected from MineCraft server logs by UBC CS research group.
    


- **Variables in players.csv**
  - **experience** (Character)  
    - Meaning: Level of player experience (e.g., Beginner, Amateur, Regular)  
    - Potential issues: These are strings, so may need to convert to factors for modeling.  
  - **hashedEmail** (Character)  
    - Meaning: This is a unique ID for each player. Both players.csv and sessions.csv have it, so we can match a player with their game sessions.  
    - Potential issues: Some players might have more than one account, which could mess up counts a bit.
  - **name** (Character)  
    - Meaning: Player's display name in the game   
  - **gender** (Character)  
    - Meaning: The player's gender. Most are "Male" or "Female", but there are also a few other options like Non-binary, Two-Spirited, Prefer not to say, etc.  
    - Potential issues: Some of the less common categories have very few players, which might make analysis tricky.
  - **played_hours** (Numeric)  
    - Description: Total hours played  
    - Min: 0 
    - Max: 223.1 
    - Mean: 5.85 
    - Median: 0.1  
  - **Age** (Numeric)  
    - Description: Age of player  
    - Min: 9  
    - Max: 58 
    - Mean: 21.14  
    - Median: 19
  - **subscribe** (Boolean)  
    - Meaning: Whether the player subscribed to the game newsletter (TRUE = subscribed, FALSE = not subscribed).  

- **Variables in sessions.csv**
  - **hashedEmail** (Character)  
    - Meaning: This is a unique ID for each player. Both players.csv and sessions.csv have it, so we can match a player with their game sessions.  
    - Potential issues: Some players might have more than one account, which could mess up counts a bit.  
  - **start_time** (Character)  
    - Meaning: The time when the play session started, like "30/06/2024 18:12".
  - **end_time** (Character)  
    - Meaning: The time when the play session ended, like "30/06/2024 19:05".
  - **original_start_time** (Double)  
    - Meaning: The numeric version of the session start time, stored as a big number in scientific notation like 1.71977E+12.
  - **original_end_time** (Double)  
    - Meaning: The numeric version of the session end time, stored as a big number in scientific notation.
  

- **Potential issues**
  - Some players may have multiple accounts, which could lead to duplicate records.
  - If the server connection was unstable, some of the playersâ€™ actions may not have been recorded.
